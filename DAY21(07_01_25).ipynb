```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
cars[['HP','WT','VOL','SP']]=sc.fit_transform(cars[['HP','WT','VOL','SP']])
```

```python
np.round(cars.describe(),2)
```

```python
ml_v=smf.ols('MPG~VOL',data = cars).fit()
ml_v.rsquared
```

```python
ml_w=smf.ols('MPG~WT',data = cars).fit()
ml_w.rsquared
```

```python
ml_v.params
```

```python
ml_w=smf.ols('MPG~WT',data = cars).fit()
ml_w.rsquared
```

```python
ml_wv=smf.ols('MPG~VOL+WT',data = cars).fit()
ml_wv.rsquared
```

```python
ml_sp=smf.ols('MPG~SP',data = cars).fit()
ml_sp.rsquared
```

```python
ml_hp=smf.ols('MPG~HP',data = cars).fit()
ml_hp.rsquared
```

```python
import statsmodels.formula.api as smf
model = smf.ols('MPG~VOL+WT+SP+HP', data=cars).fit()
```

```python
model.rsquared
```

```python
model.params
```
